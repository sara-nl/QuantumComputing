## Modifying a quantum state <a id ='modifyingastate'></a>

As we previously mentioned, the state of a qubit can be represented as a point (vector) in Bloch sphere. Different points in the sphere represent different qubit states. It is there rather intuitively that if we want to modify the state of a qubit, we need a way of moving the the point in Bloch's Sphere. Movements within a shhere can be done by rotating and reflecting the vector associated to the point. Mathematically, the rotation and reflection of a vector can be achieved by multiplying the vector by a rotation or reflection matrix. In quanutm computing a "gate" is an operation that modifies the state of a qubit. We know, that such operations are matrices. In particular in order to be consider a gate, the matrix most be unitary and have the same number of input and output qubits.
The action of a gate on a specific qubit state is then determined by multiplying the statevector $|\psi\rangle$ which represents the qubit state by the matrix $U$ representing the gate.

Let for example $U$ be a rotation around the $X$ axis, then U has the form:

$U =
\left(\begin{array}{cc}
0 & 1\\
1 & 0
\end{array}\right)
$

Let $|\psi\rangle$ be the basis state $|0\rangle$:

$|\psi\rangle =
\left(\begin{array}{c}
1\\
0
\end{array}\right)
$

Then the action of the gate $U$ on $|\psi\rangle$ is

$U | \psi \rangle =
\left(\begin{array}{cc}
0 & 1\\
1 & 0
\end{array}\right)
\left(\begin{array}{c}
1\\
0
\end{array}\right)=
\left(\begin{array}{c}
0\\
1
\end{array}\right)
$

A quantum circuit is nothing else than a sequence of operations (gates) applied on the qubit state.
They are the way we program quantum computers. When designed to solve a specific problem, then they are the graphical representations of a quantum algorithm. They are read from the left to the right. 
On the very left, the individual qubit. From there on to the right, there is a line, called a wire, that describes the evolution of the qubit during computation. On this wire, one can apply unitary operations at a certain point, by means of quantum gates. Finally, the state is measured. We will talk about this in the next section.



## Single qubit gates

### X-gate
Rotation of 180 degrees around the X-axis of Bloch sphere. It switches the amplitudes of state $|0 \rangle$ and $|1\rangle$. Classically it corresponds ot a bit flip. 
State flips over XY-plane  ($\theta$ become $\pi -\theta$) and XZ-plane ($\phi$ becomes $-\phi$)

$X =
\left(\begin{array}{cc}
0 & 1\\
1 & 0
\end{array}\right)
$

$X \left(\begin{array}{c}
1\\
0
\end{array}\right)=
\left(\begin{array}{cc}
0 & 1\\
1 & 0
\end{array}\right)
\left(\begin{array}{c}
1\\
0
\end{array}\right)=
\left(\begin{array}{c}
0\\
1
\end{array}\right)
$

### Z-gate
Rotation of 180 degrees around Z-axis. It is a phase flip. The quantum state is reflected on the XZ-plane and YZ-plane ($\phi$ becomes $\pi + \phi$). The state $|0\rangle$ does not change.

$Z =
\left(\begin{array}{cc}
1 & 0\\
0 & -1
\end{array}\right)
$

$Z \left(\begin{array}{c}
1\\
0
\end{array}\right)=
\left(\begin{array}{cc}
1 & 0\\
0 & -1
\end{array}\right)
\left(\begin{array}{c}
1\\
0
\end{array}\right)=
\left(\begin{array}{c}
1\\
0
\end{array}\right)
$

### Y-gate
Rotation of 180 degrees around Y-axis. The quantum state flips over XY-plane ($\theta$ become $\pi -\theta$) and flips on the YZ plane ($\phi$ becomes $\pi - \phi$)

$Y =
\left(\begin{array}{cc}
0 & -i\\
i & 0
\end{array}\right)
$

$Y \left(\begin{array}{c}
1\\
0
\end{array}\right)=
\left(\begin{array}{cc}
0 & -i\\
i & 0
\end{array}\right)
\left(\begin{array}{c}
1\\
0
\end{array}\right)=
\left(\begin{array}{c}
0\\
i
\end{array}\right)
$

### Hadamard gate

The Hadamard gate (H-gate) is a fundamental quantum gate. It allows us to move away from the poles of the Bloch sphere and create a superposition of $|0\rangle$ and $|1\rangle$. 

$H =
\frac{1}{\sqrt 2}\left(\begin{array}{cc}
1 & 1\\
1 & -1
\end{array}\right)
$

$H \left(\begin{array}{c}
1\\
0
\end{array}\right)=
\frac{1}{\sqrt 2}
\left(\begin{array}{cc}
1 & 1\\
1 & -1
\end{array}\right)
\left(\begin{array}{c}
1\\
0
\end{array}\right)=
\frac{1}{\sqrt 2}
\left(\begin{array}{c}
1\\
1
\end{array}\right)
$

## Parametrized gates

Require the input of a parameter to be fully defined. Examples are the P-gate and the Rx, Ry and Rz (rotation by a given angle aroudn the axes respectively).
